## Histograms

Visualise the distribution of a single continuous variable by dividing the x axis into bins and counting the number of observations in each bin. Histograms (`geom_histogram()`) display the counts with bars.

You can define the number of bins (e.g. divide the data five bins) or define the binwidth (e.g. each bin is size 10).

Distributions can be visualised as:
* counts
* scaled count
* density
* scaled density
* scaled density as a percentage

```python
diamonds.head(5)
```

If you create a basic histogram, you will be prompted to define the binwidth or number of bins.

```python
(
    ggplot(diamonds, aes(x = 'carat')) 
    + geom_histogram()
)
```

You can define the width of the bins, by specifying the `binwidth` inside `geom_histogram()`.

```python
(
    ggplot(diamonds, aes(x = 'carat')) 
    + geom_histogram(binwidth = 0.5) # specify the binwidth
)
```

Or you can define the number of bins by specifying `bins` inside `geom_histogram()`. Note, the example below uses 10 bins, however you can't see them because some of the bins are too small to be noticeable.

```python
(
    ggplot(diamonds, aes(x = 'carat')) 
    + geom_histogram(bins = 10) # specify the number of bins
)
```

There are different ways to visualise the distribution, you can specify this using the `y` argument within `aes()`. In the example below I'm using the default setting: raw count with `stat(count)`.

```python
(
    ggplot(diamonds, aes(x = 'carat'
                         , y = 'stat(count)' # specify each bin is a count
                        )) 
    + geom_histogram(binwidth = 0.50)  
)
```

You can normalise the raw count to 1 by using `stat(ncount)`:

```python
(
    ggplot(diamonds, aes(x = 'carat'
                         , y = 'stat(ncount)' # normalise the count to 1
                        )) 
    + geom_histogram(binwidth = 0.50)  
)
```

You can display the density of points in a bin, (this is scaled to integrate to 1) by using `stat(density)`:

```python
(
    ggplot(diamonds, aes(x = 'carat'
                         , y = 'stat(density)' # density
                        )) 
    + geom_histogram(binwidth = 0.50)  
)
```

The proportion of bins can be shown, in the example below the bin = 0.5 accounts for about ~55% of the data.

```python
(
    ggplot(diamonds, aes(x = 'carat'
                         , y = 'stat(width*density)') # show proportion
          ) 
    + geom_histogram(binwidth = 0.5)
)
```

We can also display counts as percentages by using the `percent_format()` which requires the `mizani.formatters` library:

```python
from mizani.formatters import percent_format 
(
    ggplot(diamonds, aes(x = 'carat', y = 'stat(width*density)')) 
    + geom_histogram(binwidth = 0.5)
    + scale_y_continuous(labels = percent_format()) # display labels as a percentage
)
```

Instead of using `stat` you can use `stat_bin` defined within `geom_histogram()`:

```python
(
    ggplot(diamonds, aes(x = 'carat'))
    + geom_histogram(y = 'stat_bin(density)', binwidth = 0.5)
)
```

You can also flip the x-y coordinates:

```python
(
    ggplot(diamonds, aes(x = 'carat', y = 'stat(density)')) 
    + geom_histogram(binwidth = 0.5)
    + coord_flip()
)
```

You can visualise counts by other variables using `fill`:

```python
(
    ggplot(diamonds, aes(x = 'carat', y = 'stat(count)',fill = 'cut')) 
    + geom_histogram(binwidth = 0.5)
)
```

You can visualise too-small-to-see bars by transforming the y-axis scaling by using `scale_y_sqrt()` square-root scale or `scale_y_log10()` for a log-scale (similarly use `scale_x_sqrt()` and `scale_x_log10()` to transform the x-axis).

```python
(
    ggplot(diamonds, aes(x = 'carat', y = 'stat(count)')) 
    + geom_histogram(binwidth = 0.5)
    + scale_y_sqrt() # square root scale
)
```

```python
(
    ggplot(diamonds, aes(x = 'carat', y = 'stat(count)')) 
    + geom_histogram(binwidth = 0.5)
    + scale_y_log10() # log scale
)
```

Change the look of your plot:

```python
(
    ggplot(diamonds, aes(x = 'carat', y = 'stat(density)'))
    + geom_histogram(binwidth = 0.5
                    , fill = 'green'  # change the fill colour
                    , colour = 'pink' # change the outline
                    , size = 2        # change the thickness of the outline
                    , alpha = 0.7     # change the transparency 
                    )
    + theme_xkcd()                    # play with themes (look/arrangement)
)
```

When faceting histograms with scaled counts/densities, they are scaled by each facet, and not overall:

```python
(
    ggplot(diamonds, aes(x = 'carat', y = 'stat(ncount)')) 
    + geom_histogram(binwidth = 0.5)
    + facet_wrap('color')
)
```

```python
(
    ggplot(diamonds, aes(x = 'carat', y = 'stat(ncount)')) 
    + geom_histogram(binwidth = 0.5)
    + facet_grid('cut ~ color')
)
```
